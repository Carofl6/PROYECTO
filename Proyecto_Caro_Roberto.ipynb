> Dentro de la empresa Ciasa Comercial S.A. de C.V. se realizó un mapeo del proceso de compras y resulto de la siguiente manera:

* Si hay cambios regresa a dirección para la aprobación, de lo contrario, se le da un anticipo del 30% al proveedor dentro de las primeras dos semanas de tener la orden emitida.
